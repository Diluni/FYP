```python
import spacy
from spacy.matcher import Matcher
from spacy.tokens import Span
from spacy import displacy
import pandas as pd
from spacy.matcher import PhraseMatcher
import nltk
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
# importing datasets
df = pd.read_csv("D:\FYP\Entities\Excel_files\Text_db_final_4_replacing n with period.csv", encoding = 'utf-8')
jobtitledf = pd.read_csv("D:\FYP\Entities\Job Title\jobtitles.csv")
jobtitleESCOdf = pd.read_csv("D:\FYP\Entities\Job Title\occupations_en modified.csv")
```

```python
# converting df cols to string
jobtitleESCOdf.preferredLabel = jobtitleESCOdf.preferredLabel.astype(str)
jobtitleESCOdf.altLabels = jobtitleESCOdf.altLabels.astype(str)
jobtitleESCOdf.hiddenLabels = jobtitleESCOdf.hiddenLabels.astype(str)
df.text = df.text.astype(str)
df.name = df.name.astype(str)
```

```python
# splitting jobtitleESCOdf.altLabels in to multiple lines
k = jobtitleESCOdf.altLabels.str.split('\n', expand = False)
kwlist = []
for i in range(len(k)):
    for j in range(len(k[i])):
        kwlist.append(k[i][j])
```

```python
# converting dataframe columns to lists
preferredLabel = jobtitleESCOdf.preferredLabel.tolist()
keywordlist = jobtitledf['title'].tolist()
```

```python
# putting all  lists to one list
keywordlist.extend(kwlist)
keywordlist.extend(preferredLabel)
```

```python
matcher = PhraseMatcher(nlp.vocab, attr = 'LOWER')
```

```python
patterns = [nlp.make_doc(text) for text in keywordlist]
```

```python
matcher.add("Jobtitles",patterns)
```

```python
extracted = {}
notextracted = {}
```

```python
def jobTitleExtracter(index):
    print(index)
    t = df.text[index]
    id_ = df.name[index]
    doc = nlp(t)
    k = matcher(doc)
    try:
        titlelist = []
        if len(k) == 1:
            span = doc[k[0][1]:k[0][2]]
            titlelist.append(span.text)
            extracted[id_] = titlelist
            #print(span.sent)
        elif len(k) > 1:
            #print(len(k))
            #agelist = []
            for i in range(len(k)):
                span = doc[k[i][1]:k[i][2]]
                #print(span.sent)
                titlelist.append(span.text)
            extracted[id_] = titlelist
        else:
            print("No title")
            notextracted[id_] = None
    except IndexError:
        print("No title")
        notextracted[id_] = ""
```

```python
for i in range(len(df.text)):
    jobTitleExtracter(i)
```

```python
from nltk.stem import PorterStemmer
from nltk.stem import LancasterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
df["textlemma"] = df['text'].apply(lambda x: " ".join([y.lemma_ for y in nlp(x)]))
```

```python
df_fulllist['kw_lemma'] = df_fulllist['kw'].apply(lambda x: " ".join([y.lemma_ for y in nlp(x)]))
```

```python
#create an object of class PorterStemmer
porter_stemmer = PorterStemmer()
lancaster=LancasterStemmer()
```

```python
def stem_sentencesPorter(sentence):
    tokens = sentence.split()
    stemmed_tokens = [porter_stemmer.stem(token) for token in tokens]
    return ' '.join(stemmed_tokens)
```

```python
df_fulllist['kw_StemPort']= df_fulllist['kw'].apply(stem_sentencesPorter)
```

```python
df.textPort = df.text.apply(stem_sentencesPorter)
```

```python
def stem_sentencesLan(sentence):
    tokens = sentence.split()
    stemmed_tokens = [lancaster.stem(token) for token in tokens]
    return ' '.join(stemmed_tokens)
```

```python
df_fulllist['kw_StemL']= df_fulllist['kw'].apply(stem_sentencesLan)
```

```python
df.textL = df.text.apply(stem_sentencesLan)
```
